```python

import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.head()
```

```python
#creating a helper function to get first 3 cast members from cast column for each movie
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
        
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#overview column is a string hence we need to convert it a list
movies['overview'][0]
```

```python
#this is used to convert string into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing spaces 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#joining all the columns created into tags column
movies['tags'] = movies['genres']+movies['cast']+movies['crew']+movies['keywords']
```

```python
#creating a new dataframe 
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# Keep only the first occurrence of each column (this removes the duplicate)
new_df = new_df.loc[:, ~new_df.columns.duplicated()]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.fit_transform(new_df['tags']).toarray().shape #tells that we have 4806 movies and we have converted it to vector wrt to5000 common words
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loving')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)#used to calculate cosinesimilarity between vectors vreated by textvectorization
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6] #creates a lsit of tuples where we obtain index position and cosine simlarity of first movie with all other 4806 movies
#then we did sorting by using sorted(list(enumerate(similarity[0]))
#after this we obtain list sorted in descending order of cosine similarity using sorted(list(enumerate(similarity[0])),reverse = True)
#but problem in above is that sorting done according to index position in descending order we need to sort  cosine similairty values hence
# sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])  we use this code.
#now we require only first five movies hence sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6] used
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"]== movie].index[0] #this code fetches the index position for avatar movie and stores in movie index
    distances = similarity[movie_index]#calculating distances for the movie index provided
    movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
